## apply()
데이터 프레임에서 Series 객체를 통해 평가하고자하는 Python 로직을 지정할 수있게 해줍니다. 
```
각 숫자 열의 표준 편차 계산

영화 칼럼의 모든 영화 이름 소문자
```

apply () 메서드를 사용하려면 각 Series 객체에 적용 할 벡터화 된 연산을 전달해야합니다. 이 메소드는 기본적으로 데이터 프레임의 컬럼을 통해 실행되지만, 이것을 변경하기 위해 axis 매개 변수를 사용할 수 있습니다 (나중에 다룰 것입니다). 벡터화 된 연산이 대개 단일 값을 반환하면 각 열에 대해 계산 된 값이 들어있는 Series 객체를 반환합니다. 예를 들어 다음 코드에서 람다 함수를 사용하여 각 float 열에 2를 곱합니다. <br>
from [dataquest 중](https://www.dataquest.io/m/79/pandas-internals%3A-dataframes/5/apply-logic-over-the-columns-in-a-dataframe)

````python
types = fandango_films.dtypes
print(types)
````

````python
float_columns = types[types.values == 'float64'].index
print(float_columns)
````

````python
# float_df contains only the float columns
float_df = fandango_films[float_columns]
float_df.head()
````

````python
# usage of a lambda function
float_df.apply(lambda x: x*2)
````

## lambda()

일반적으로 각 요소의 값을 2로 나누거나 곱하는 등의 값을 반환하면 모든 값을 변환하여 새 데이터 프레임으로 반환합니다.

````python
import numpy as np
````

````python
float_df.apply(lambda x: np.std(x))
````


NumPy는 빠른 벡터화 연산을 기반으로합니다. Pandas는 DataFrame.apply () 메소드를 사용할 때 빠른 벡터화를 활용할 수 있습니다. <br>
from [dataquest](https://www.dataquest.io/m/79/pandas-internals%3A-dataframes/5/apply-logic-over-the-columns-in-a-dataframe)

````python
halved_df = float_df.apply(lambda x: x/2)
halved_df.head()
````

pandas가 Series 객체로 취급하는 데이터 프레임의 행에 함수를 적용하려면 축 매개 변수를 1로 설정해야합니다. 행을 적용하면 각 동영상에 대한 여러 열 값의 표준 편차를 계산할 수 있습니다 :

````python
rt_mt_user = float_df[['RT_user_norm', 'Metacritic_user_nom']]
rt_mt_user.apply(lambda x: np.std(x), axis=1)
````

위의 코드는 데이터 프레임을 원하는 두 열로 필터링합니다. std ()는 각 반복에 대한 값을 반환하기 때문에 RT_user_norm 및 Metacritic_user_nom의 각 영화 등급에 대한 표준 편차를 포함하는 Series 객체를 반환합니다.

````python
rt_mt_means = float_df[['RT_user_norm', 'Metacritic_user_nom']]
````

````python
rt_mt_user = float_df[['RT_user_norm', 'Metacritic_user_nom']]
rt_mt_means = rt_mt_user.apply(lambda x: np.mean(x), axis=1)
print(rt_mt_means[0:5])
````
