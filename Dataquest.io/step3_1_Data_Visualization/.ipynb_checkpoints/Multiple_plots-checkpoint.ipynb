#### First, 
Understanding what pyplot was automatically storing under the hood when we create a single plot.

a countainer for all plots was created(returned as a Figure object)

a container for the plot was positioned on a grid (the plot returned as an Axes object) visual symbols were added to the plot(using the Axes mehtods) by 90 degrees.

모든 플롯에 대한 컨테이너가 작성되었다 -> Figure 객체로 반환 됨. 
플롯을 위한 컨터이너?는 그리드 상에 배치되었다. -> 플롯은 Axes 객체로 반환된다.
시각적 기호가 플롯에 추가되었다. -> Axes 방법을 사용 

#### <p style="color:skyblue">Figure</p>
* Figure는 모든 plot의 컨테이너 역할을 한다. 
* 컨테이너 내의 plot에 대한 모양과 동작을 커스터마이징할 수 있는 메소드를 가진다.
* pyplot 함수를 호출하는 대신 반환 값을 변수에 지정.
* Figure가 만들어졌으면, Figure context 안에서 데이터가 없는 a single plot에 대한 an Axes가 만들어진다. 
-- 데이터 없이 랜더링 되면, plot은 빈 plot과 유사하다.
* Axes 객체는 다음과 같이 plot의 다양한 구성 요소에 대한 자체 컨테이너 역할을 한다. 
    * values on the x-axis and y-axis
    * ticks on the x-axis and y-axis
    * all visual symbols
        * markers. lines, gridlines
        
        
* Plot은 AXES 클래스의 인스턴스를 사용해 표현되기도하는데, matplotlib의 subplots으로 지칭된다. 이때는 add_subplot을 사용


```python
fig = plt.figure()
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax2 = fig.add_subplot(3, 1, 3)
plt.show()
```

# resizing
matplotlib used the default dimensions for the total plotting area instead of resizing it to accommodate the plots. 

If we want to expand the plotting rea, we have to specify this ourselves when we create the figure.

\# figsize #plt.figure()

The unit for both width and height values is inches.

```python
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

# Expanding the plotting area horizontally

we'd need to increase the spacing between the plots to keep them from overlapping. Expanding the plotting area horizontally improved the readability of the x-axis tick labels and minimized the amount of space between the 2 line charts.

Let's visualize data from a few more years to see if we find any evidence for seasonality between those years.

```python
fig = plt.figure(figsize=(12, 12))

for i in range(5):
    ax = fig.add_subplot(5, 1, i+1)
    start_index = i * 12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])
plt.show()
```

# Legend
to help remind us which year each line corresponds to, we can add a legend that links each color to the year the line is representing.

\# plt.legend()   #Axes.plot()    #Axes.legend()

```python
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c = 'red', label='1949')
plt.legend(loc='upper left')
```

```python
fig = plt.figure(figsize=(10, 6))

# 1948 ~ # 1952
col = ['red', 'blue', 'green', 'orange', 'black']

#legend
unrate['YEAR'] = unrate['DATE'].dt.year
years_label = list(pd.unique(unrate['YEAR']))
for i in range(5):
   # ax = fig.add_subplot(5, 1, i+1)
    start_index = i * 12
    end_index = (i+1) * 12
    # col
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c = col[i], label = years_label[i])
    plt.legend(loc='upper left')
plt.show()    
```

```python
fig = plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
#     ax = fig.add_subplot(5, 1, i+1)
    start_index = i *12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset['MONTH'], subset['VALUE'], c = colors[i], label=label)
    plt.legend(loc='upper left')
    
```

```python
fig = plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
#     ax = fig.add_subplot(5, 1, i+1)
    start_index = i *12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset['MONTH'], subset['VALUE'], c = colors[i], label=label)
    plt.legend(loc='upper left')
    
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
