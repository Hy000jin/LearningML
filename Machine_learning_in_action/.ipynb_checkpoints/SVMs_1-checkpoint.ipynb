## 간략한 형태의 알고리즘

```python
dataMatrix = mat(dataArr)
# print(dataMatrix[:10])
labelMat = mat(labelArr).transpose()
# labelMat = mat(labelArr)
# print(labelMat[:10])

# alpha 벡터 생성, b의 초기화
b = 0; m, n = shape(dataMatrix) # 데이터 shape
# instance 100, feature 2개
alphas_num = mat(zeros((m, 1)))
iter_num = 0
# print(m, n)
print(shape(alphas_num))
print(alphas_num[:10])
```

# 연습

```python
maxIter = 40
iter = 0.001
C = 0.6
toler = 0.1

while (iter < maxIter):
    alphaPairsChanged = 0
    for i in range(m):
        print("\n===========")
        print("alphaPairsChanged: ", alphaPairsChanged)
        print("i: ", i)
        fxi = float(multiply(alphas_num, labelMat).T *
                   (dataMatrix*dataMatrix[i, :].T)) + b
    #     print((multiply(alphas_num, labelMat).T))
        print("분류항목에 대한 우리의 예측인 fxi를 계산한다.\n fxi: ", fxi)
    #     print(dataMatrix[i, :].T)
    #     print(dataMatrix[:10])
    #     print(shape(dataMatrix[i, :].T))
    #     print((dataMatrix*dataMatrix[i, :].T)[:10])
    #     print(fxi)
    #     break;
        Ei = fxi - float(labelMat[i])
    #     print(Ei*labelMat[i])
        print("오류 Ei는 예측과 이 사례ㅢ 실제 분류 항목을 기반으로 계산된다. \n Ei: ", Ei)

        # toler 오차범위 0.1 C:상수 3
        if ((labelMat[i]*Ei < toler) and (alphas_num[i] < C)) or ((labelMat[i]*Ei > toler) and (alphas_num[i] > 0)):
            print("양의 마진과 음의 마진을 둘 다 검토")
        # 조건식 충족 
            j = i
            while (j == i):
                j = int(random.uniform(i, m))
                print("임의의 j: ", j)

            print("alphas_num[i]: ", alphas_num[i])
            print("alphas_num[j]: ", alphas_num[j])
            fxj = float(multiply(alphas_num, labelMat).T *
               (dataMatrix*dataMatrix[j, :].T)) + b
            print("예측 fxj: ", fxj)
            Ej = fxj - float(labelMat[j])
            print("오류 Ej: ", Ej)
            alphaIold = alphas_num[i].copy()
            alphaJold = alphas_num[j].copy()
            
            if (labelMat[i] != labelMat[j]):
                print("알파는 0 또는 C와 동일하지 않다. L과 H는?")
                L = max(0, alphas_num[j] - alphas_num[i])
                H = min(C, C + alphas_num[j] - alphas_num[i]) 
    #             break;
            else:
                L = max(0, alphas_num[j] + alphas_num[i] - 1)
                H = min(C, alphas_num[j] + alphas_num[i])
            print("\nL: ", L)
            print("H: ", H)    
            if L == H: 
                print("L==H"); 
                continue
            eta = 2.0 * dataMatrix[i, :] * dataMatrix[j, :].T - \
                dataMatrix[i, :] * dataMatrix[i, :].T - \
                dataMatrix[j, :] * dataMatrix[j, :].T
    #         print(eta)
    #         break;

            if eta >= 0:
                print("\neta는 alpah[j]를 변경하기 위한 최적화된 계수")
                print("eta > 0")
                continue
            alphas_num[j] -= labelMat[j]*(Ei - Ej) / eta
    #         alpahs_num[j] = clipAlpha(alphas_num[j], H, L)
            if alphas_num[j] > H:
                alphas_num[j] = H
            if L > alphas_num[j]:
                alphas_num[j] = L
            print("최적화된 계수를 거친 alphas_num[j]: ", alphas_num[j])
            if (abs(alphas_num[j] - alphaJold) < 0.00001):
                print("j not moving enough"); continue
                
            alphas_num[i] += labelMat[j] * labelMat[i] * (alphaJold - alphas_num[j])
            print("\n반대쪽에 있는 j와 동일하게 계산된 값으로 j를 갱신한다.")
            print("갱신된 alphas[i]", alphas_num[i])
            b1 = b - Ei - labelMat[i] * (alphas_num[i] - alphaIold) * dataMatrix[i, :] * dataMatrix[i, :].T - labelMat[j] * (alphas_num[j] - alphaJold) * dataMatrix[i, :] * dataMatrix[j, :].T
            b2 = b - Ej - labelMat[i] * (alphas_num[i] - alphaIold) * dataMatrix[i, :] * dataMatrix[i, :].T - labelMat[j] * (alphas_num[j] - alphaJold) * dataMatrix[j, :] * dataMatrix[j, :].T
            print("\nb1: ", b1)
            print("b2:", b2)
            break;
    #         if (0 < alphas_num[i])
            if (0 < alphas_num[i]) and (C > alphas_num[i]):
                b = b1
            elif (0 < alphas_num[j]) and (C > alphas_num[j]):
                b = b2
            else:
                b = (b1 + b2) / 2.0
            alphaPairsChanged += 1
            print("iter: %d i:%d, pairs changed %d") % (iter, i, alphaPairsChanged)
        
        if (alphaPairsChanged == 0): 
            iter +=1
        else:
            iter = 0
#         print("iteration number:  %d") % iter
        break;

print(b)
print(alphas_num)
print("==end==\n")
    
        
```
